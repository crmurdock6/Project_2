### Load dataframe of stadiums with longitudes and latitudes and use plotly to make ineractive graph

```python
path = '/Users/murdock/Documents/metis/project2_luther/data-visualization/ncaa-football-stadiums/data/'
college_coords = pd.read_csv(path + 'stadiums-geocoded.csv')
```

```python
college_coords.head()
```

```python
latitudes = college_coords['latitude']
longitudes = college_coords['longitude']
names = college_coords['stadium']
```

```python
college_coords.sort_values('capacity', ascending=True)
```

```python
conf = college_coords['conference'].unique()
conf
```

```python
limits = [(0,20000),(20001,40000),(40001,60000),(60001,80000),(80001,100000)]
college_coords['text'] = college_coords['team'] + '<br>Capacity ' + (college_coords['capacity']).astype(str)
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(246, 52, 13)","rgb(246, 161, 202)","rgb(53, 255, 147)","rgb(153, 46, 165)","rgb(140, 23, 74)","rgb(221, 147, 82)","rgb(205, 237, 130)","rgb(40, 237, 204)","rgb(108, 112, 111)","rgb(127, 5, 7)","rgb(239, 107, 110)","rgb(195, 252, 194)","rgb(197, 171, 198)","rgb(171, 173, 198)","rgb(171, 198, 186)","rgb(11, 102, 61)","rgb(229, 209, 149)","rgb(237, 252, 151)","rgb(27, 249, 220)","rgb(237, 80, 221)","rgb(140, 74, 77)"]
teams = []
scale = 5000

for i in range(len(conf)):
    lim = conf[i]
    #print(lim)
    df_sub = college_coords[college_coords['conference'] == lim]
    #print(df_sub)
    team = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['capacity']/scale * 8,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = conf[i] )
    teams.append(team)


# for k, v in teams[1].items():
#    print(k, v)

layout = dict(
        title = 'NCAA Div I Football Stadiums',
        showlegend = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict( data=teams, layout=layout )
plotly.offline.iplot( fig, validate=False, filename='NCAA Stadium Locations' )
#py.iplot( fig, validate=False, filename='NCAA Stadium Locations' )
```
